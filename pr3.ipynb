```python
import pandas as pd
import matplotlib.pyplot as plt

 
data = {
    'Singer': ['Arijit Singh', 'Shreya Ghoshal', 'Sonu Nigam', 'Lata Mangeshkar', 'Kishore Kumar'],
    'Years_Active': [15, 20, 30, 70, 40],
    'Number_of_Songs': [500, 700, 600, 1000, 800],
    'Awards_Won': [50, 60, 40, 100, 80]
}

#  a DataFrame
df = pd.DataFrame(data)

# Step 3: Plot a bar chart for Number of Songs
plt.figure(figsize=(10, 5))
plt.bar(df['Singer'], df['Number_of_Songs'], color=['blue', 'green', 'red', 'purple', 'orange'], edgecolor='black')

 
plt.xlabel('Singers', fontsize=12)
plt.ylabel('Number of Songs', fontsize=12)
plt.title('Number of Songs by Each Singer', fontsize=14)
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.7)

#  Show the plot
plt.show()
 
```
